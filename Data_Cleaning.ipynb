```python
# 1. Import libraries

from sqlalchemy import create_engine as ce
import zipfile
import pandas as pd

# open zipped voyages dataset
with zipfile.ZipFile("voyages_202301.csv.zip") as z:
   # open the csv file in the dataset
   with z.open("voyages_202301.csv") as f:
       
      # read the dataset
        voyages_df = pd.read_csv(f)

        
#read in vessels dataset
vessels_df = pd.read_csv('fishing-vessels-v2.csv')


#read in anchorages dataset
anchorages_df = pd.read_csv('named_anchorages_v2_20221206.csv')

```

```python
voyages_dfClean = pd.DataFrame(voyages_df[["mmsi","trip_start_anchorage_s2id","trip_end_anchorage_s2id"]])
voyages_dfClean
```

```python
vessels_dfClean = pd.DataFrame(vessels_df[["mmsi","flag_gfw","vessel_class_gfw","length_m_gfw","tonnage_gt_gfw","fishing_hours_2018","fishing_hours_2019","fishing_hours_2020"]])
vessels_dfClean
```

```python
anchorages_dfClean = pd.DataFrame(anchorages_df[["s2id","lat","lon","label","sublabel","iso3"]])
anchorages_dfClean
```

```python
vessels_dfClean["fishing_hours_2018"].fillna(0,inplace = True)
vessels_dfClean
```

```python
vessels_dfClean["fishing_hours_2019"].fillna(0,inplace = True)
vessels_dfClean
```

```python
vessels_dfClean["fishing_hours_2020"].fillna(0,inplace = True)
vessels_dfClean
```

```python
mergedvoyagesvessels= pd.merge(vessels_dfClean,voyages_dfClean, how='inner', on='mmsi')
mergedvoyagesvessels
```

```python
voyagesvessels= mergedvoyagesvessels[['mmsi', 'flag_gfw','trip_start_anchorage_s2id', 'trip_end_anchorage_s2id']]
voyagesvessels
```

```python
voyagesanchorages_df= pd.merge(voyagesvessels, anchorages_dfClean,left_on='trip_start_anchorage_s2id', right_on='s2id')
voyagesanchorages_df
```

```python
voyagesanchorages_df.rename(columns= {'lat':'lat_start', 'lon':'lon_start'},inplace=True)
voyagesanchorages_df
```

```python
finaltripend_df= pd.merge(voyagesanchorages_df, anchorages_dfClean,left_on='trip_end_anchorage_s2id', right_on='s2id')
finaltripend_df
```

```python
finaltripend_df.rename(columns={'lat':'lat_end','lon':'lon_end',}, inplace=True)
```

```python
finaltripend_df.drop(['s2id_x','label_x','sublabel_x', 'iso3_x','s2id_y','label_y','sublabel_y','iso3_y'], axis=1, inplace=True)
finaltripend_df
```

```python
finaltripend_df.to_csv('Finalcleanedvoyages.csv', index=False)
```
